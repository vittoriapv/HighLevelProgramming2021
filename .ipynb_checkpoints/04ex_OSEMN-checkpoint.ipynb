4\. load the binary file named *credit_card.dat* and convert the data into the real credit-card number.
Each line correspond to a credit card number.
Each character is composed by 6 bit (even the space) and the last 4 bit are just a padding

```python
file_name = "C:\\Users\\TEST\\LaboratoryOfComputationalPhysics\\credit_card.dat"
card_list = []
with open(file_name, mode='r') as f:
    for line in f:
        cardnum=''
        for i in [i*6 for i in range(int((len(line)-4)/6))]:
            y = line[i:i+6]
            cardnum+=chr(int(y,2))
        if cardnum == '':#to remove any cases if the cardnum can be empty
            continue
        card_list.append(cardnum)
f.close()
print(card_list)
```

5\. Load the file "user_data.json", filter the data by the "CreditCardType" field equals to "American Express". Than save the data a to CSV.

```python
import json
data = json.load(open('user_data.json'))
#print(data)
only_american = []
for i in range(len(data)):
    if data[i]['CreditCardType']=='American Express':
        only_american.append(data[i])
print(only_american)

import csv
filename = 'C:\\Users\\TEST\\LaboratoryOfComputationalPhysics\\data\\American_express.csv'

with open(filename, mode='w+') as csv_file:
    fieldnames = ['ID', 'JobTitle', 'EmailAddress','FirstNameLastName','CreditCard','CreditCardType']
    writer = csv.DictWriter(csv_file, fieldnames=fieldnames)

    writer.writeheader()
    for i in range(len(only_american)):
        writer.writerow(only_american[i])
csv_file.close()    
```
